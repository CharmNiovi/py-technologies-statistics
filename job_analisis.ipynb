# Python vacancies hard skills analysis

### import necessary modules

```python
import pandas as pd
import matplotlib.pyplot as plt
from settings import DB
```

# Make pandas dataframe

```python
sql = """
SELECT 
    hs.tech_words,
    hs_c.word_count,
    hs_c.parse_date
FROM hard_skill_count as hs_c
JOIN hard_skills as hs 
    ON hs_c.tech_word_id=hs.id 
"""
df = pd.read_sql(sql, DB)
```

# 

```python
last_date_df = df[df["parse_date"] == max(df["parse_date"])].sort_values("word_count", ascending=False).head(15)
plt.figure(figsize=(12, 6))
plt.bar(last_date_df["tech_words"], last_date_df["word_count"])
plt.xticks(rotation=45)
plt.title("Most popular technologies")
plt.xlabel("Technologies")
plt.ylabel("Popularity")
plt.show()
```
